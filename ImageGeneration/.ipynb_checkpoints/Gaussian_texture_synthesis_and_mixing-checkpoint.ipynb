# Gaussian textures

Synthesizing new images using Gaussian models is probably the most basic idea in image synthesis. However, even if it is simple, this framework is already very rich and allows many nice recipes, going from texture synthesis to texture inpainting and texture mixing. The general model is simple and explicit, but has also many subtelties, when dealing for instance with the image boundaries or color images. 

This practical session shows how to obtain texton and Gaussian synthesis of color and gray textures, as well as texture mixing.

## 1. Periodic plus smooth image decomposition

Write a function that takes as input a gray-level image u and then computes its periodic and smooth components (Ingredient 2.3 Periodic plus smooth image decomposition).

```python
def ppluss(U):

    n,m = U.shape
    U = U.astype(np.float64)

    #Compute the derivative across the boundary.

    B = np.zeros_like(U)
    B[:,0] = U[:,-1] - U[:,0]
    B[:,-1] = U[:,0] - U[:,-1]
    B[0,:] = U[-1,:] - U[0,:]
    B[-1,:] = U[0,:] - U[-1,:]

    #Compute the Laplacian filter over the Fourier domain.

    q = np.arange(n).reshape(n, 1).astype(B.dtype)
    r = np.arange(m).reshape(1, m).astype(B.dtype)
    L = (2*np.cos( np.divide((2*np.pi*q), n) ) + 2*np.cos( np.divide((2*np.pi*r), m) ) - 4)

    #Compute the smooth part S by solving Laplacian(S)=A. This is easily solved over the Fourier domain.

    S = np.fft.fft2(B)
    S = np.divide(S, L, out=np.zeros_like(S), where=L!=0)
    S[0, 0] = 0
    S = np.real(np.fft.ifft2(S))

    #Compute the periodic part P.

    P = U - S

    return P,S
```

```python
P,S = ppluss(u_gray)
plt.imshow(P,cmap='gray',vmin=0,vmax=1)
```

## 2. Gray-level image : texton and Gaussian synthesis
1. Take a gray-level image $u_0$ of size $n_h × n_w$. Replace it by its periodic component. Compute its empirical mean $m_{u_0}$ and remove it from the image to have a zero-mean image $u_0$.

```python
m, n= u_gray.shape
u_gray_periodic, S = ppluss(u_gray)
mean_u_gray_periodic = np.mean(u_gray_periodic)
u_gray_periodic -= mean_u_gray_periodic
```

2. Then compute and visualize the texton $T(u_0)$ of $u_0$, placing 0 at the center of the image. To better see the concentration of the texton, plot the signals $x  \rightarrow T(u_0)(x,0)$ and $y \rightarrow T(u_0)(0,y)$, that are two slices of the texton image passing through the origin.

```python
# Texton of u_gray_periodic
Texton = np.fft.ifft2(np.abs(np.fft.fft2(u_gray_periodic)))
STI = np.fft.fftshift(Texton)

real_STI = np.real(STI)
plt.imshow(real_STI, cmap='gray',vmin=0,vmax=1)

# Plot slices
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
axes[0].plot( np.arange(m)-m/2, real_STI[:,int(n/2)])
axes[0].set_title('Signal x -> T(u0)(x,0)')
axes[1].plot( np.arange(n)-n/2, real_STI[int(m/2),:])
axes[1].set_title('Signal y -> T(u0)(0,y)')
fig.tight_layout()
plt.figure()
```

3. Finally, generate from $u_0$ a Gaussian texture of size $2n_h × 2n_w$.

```python
# Generate texture of same size
W= np.fft.fft2(np.random.randn(m,n))/np.sqrt(m*n)
u_hat = np.fft.fft2(Texton)*W
Gaussian_texture = np.real(np.fft.ifft2(u_hat))
Gaussian_texture += mean_u_gray_periodic

# Generate texture of 2n x 2m
Texton2= np.zeros((2*m,2*n)).astype(Texton.dtype)
Texton2[int(m/2):int(3*m/2),int(n/2):int(3*n/2)]=  Texton #real_STI - mean_u_gray_periodic 

W2= np.fft.fft2(np.random.randn(2*m,2*n))/np.sqrt(4*m*n)
u_hat_2 = np.fft.fft2(Texton2)*W2
Gaussian_texture_2 = np.real(np.fft.ifft2(u_hat_2))
Gaussian_texture_2 += mean_u_gray_periodic

# Plot both Textures
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
axes[0].imshow(Gaussian_texture, cmap='gray',vmin=0,vmax=1)
axes[0].set_title('Gaussian texture of size n x m')
axes[1].imshow(Gaussian_texture_2, cmap='gray',vmin=0,vmax=1)
axes[1].set_title('Gaussian texture of size 2*n x 2*m')
fig.tight_layout()
plt.figure()
```

## 3. Color image : texton and Gaussian synthesis

1. Take a color image $u_0$ of size $n_h × n_w$. Replace it by its periodic component (i.e. replace each channel by its periodic component). Compute its empirical mean $mu_0 \in \mathbb{R}^3$ and remove it from the image (i.e. remove to each channel its mean) to have a zero-mean color image $u_0$.

```python
m, n, c = u.shape
u_periodic = np.zeros_like(u)
mean_u_periodic = np.zeros(c)

for i in range(c):
    u_periodic[:,:,i], _ = ppluss(u[:,:,i])
    mean_u_periodic[i] = np.mean(u[:,:,i])

u_periodic -= mean_u_periodic
```

2. Then compute and visualize its color texton image $T_\alpha(u_0)$ for instance with $\alpha = (0.33, 0.5, 0.17)$ (the luminance channel), placing 0 at the center of the image.

```python
alpha=[0.33, 0.5, 0.17]
alphaTexton = np.zeros_like(u_periodic)
alphaTexton = alpha[0]*u_periodic[:,:,0] + alpha[1]*u_periodic[:,:,1] + alpha[2]*u_periodic[:,:,2]

STI_color = np.zeros_like(u_periodic)
oppPhase = np.conj(np.fft.fft2(alphaTexton))/np.abs(np.fft.fft2(alphaTexton))

for i in range(c):
    STI_color[:,:,i]= np.fft.fftshift( np.fft.ifft2(np.fft.fft2(u_periodic[:,:,i])*oppPhase))
```

3. Finally, generate from $u_0$ a Gaussian color texture of size $n_h × n_w$

```python
# Generate texture of same size
Gaussian_texture_color = np.zeros_like(u)
W= np.fft.fft2(np.random.randn(m,n))/np.sqrt(m*n)

for i in range(c):
    u_hat_color = np.fft.fft2(u_periodic[:,:,i])*W
    Gaussian_texture_color[:,:,i] = np.real(np.fft.ifft2(u_hat_color)) 
    Gaussian_texture_color[:,:,i] += mean_u_periodic[i]

plt.imshow(Gaussian_texture_color)
```

## 4. Gaussian texture mixing
1. Take two gray-level images $u_0$ and $u_1$ of same size $n_h × n_w$. Compute their mean, remove it and replace each image by its periodic component. Then generate a
sequence of $n_\rho + 1$ (with $n_\rho$ = 5 for instance) texture images from the Wasserstein
barycenter Gaussian texture model $GT(u_\rho)$, for $\rho = 0, \frac{1}{n_p} , \frac{2}{n_p} , . . . , 1$ and where $u_\rho$ is defined in Proposition 4.1.3.


```python
# Read images
u_0_gray = np.copy(u_gray)

input_image_name = "bois.png"
u_1 = plt.imread(input_image_name)

if u.shape[2]==4:
    print('removing alpha chanel')
    u_1 = u_1[:,:,:3]

u_1_gray =  0.21*u_1[:,:,0] + 0.72*u_1[:,:,1] + 0.07*u_1[:,:,2]

# verify sizes
print(u_1_gray.shape)
print(u_0_gray.shape)
nh, nw = u_1_gray.shape

# Compute mean 
mean_u_0_gray = np.mean(u_0_gray)
mean_u_1_gray = np.mean(u_1_gray)

# substract mean
u_0_gray -= mean_u_0_gray
u_1_gray -= mean_u_1_gray

# Replace images for its periodic Component
u_0_gray, _ = ppluss(u_0_gray)
u_1_gray, _ = ppluss(u_1_gray)

# Generate textons
Texton0 = np.fft.ifft2(np.abs(np.fft.fft2(u_0_gray)))
Texton1 = np.fft.ifft2(np.abs(np.fft.fft2(u_1_gray)))

for r in np.linspace(0,1,5):
    Trho = (1-r)*Texton0 + r*Texton1

# Generate Gaussian texture from mix Texton
W= np.fft.fft2(np.random.randn(nh,nw))/np.sqrt(nh*nw)
u_hat_mixture = np.fft.fft2(Trho)*W
Gaussian_texture_mixture = np.real(np.fft.ifft2(u_hat_mixture))
mean_Gaussian_texture_mixture = np.mean(Gaussian_texture_mixture)
Gaussian_texture_mixture +=  mean_Gaussian_texture_mixture # not sure this step

# Plot original textures and mixture
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 10))
axes[0].imshow(u_0_gray + mean_u_0_gray, cmap='gray',vmin=0,vmax=1)
axes[0].set_title('Texture 0')
axes[1].imshow(u_1_gray + mean_u_1_gray, cmap='gray',vmin=0,vmax=1)
axes[1].set_title('Texture 1')
axes[2].imshow(Gaussian_texture_mixture, cmap='gray',vmin=0,vmax=1)
axes[2].set_title('Gaussian texture Mixture')
fig.tight_layout()
plt.figure()
```

2. Do the same thing with two color images, using now the formulas of Proposition 4.1.4.

```python
# Read image (u_1 already loaded in cell above)
u_0 = np.copy(u)

# verify sizes
print(u_1.shape)
print(u_0.shape)
nh, nw, c = u_1.shape

# Compute mean 
mean_u_0 = np.mean(u_0, axis=(0,1))
mean_u_1 = np.mean(u_1, axis=(0,1))

# substract mean
u_0 -= mean_u_0
u_1 -= mean_u_1

# Replace images for its periodic Component
F0 = np.zeros((nh,nw,c)).astype(complex)
F1 = np.zeros((nh,nw,c)).astype(complex)
FG = np.zeros((nh,nw,c)).astype(complex)

for i in range(c):
    u_0[:,:,i], _ = ppluss(u_0[:,:,i])
    u_1[:,:,i], _ = ppluss(u_1[:,:,i])
    F0[:,:,i] = np.fft.fft2(u_0[:,:,i])
    F1[:,:,i] = np.fft.fft2(u_1[:,:,i])

# Generate textons
rho = np.linspace(0,1,5)
Grho=np.zeros((nh,nw,c,len(rho)))
Grho_list = list()
FMrho=np.zeros((nh,nw,c,len(rho)))

CR= np.conj(F0[:,:,0])*F1[:,:,0] + np.conj(F0[:,:,1])*F1[:,:,1] + np.conj(F0[:,:,2])*F1[:,:,2]
CR=CR/np.abs(CR)

for i in range(c):
    FG[:,:,i]=F0[:,:,i]*CR

# Generate texture of same size
W= np.fft.fft2(np.random.randn(nh,nw))/np.sqrt(nh*nw)

for r in range(len(rho)):
    for i in range(c):
        u_hat_mix_color = ((1-rho[r])*F1[:,:,i] + rho[r]*FG[:,:,i])*W
        Grho[:,:,i, r] = np.real(np.fft.ifft2(u_hat_mix_color)) 
    Grho_list.append(Grho[:,:,:,r])
```
